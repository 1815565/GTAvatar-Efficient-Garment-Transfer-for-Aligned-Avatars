```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

mask = cv2.imread("/home/ljr/Downloads/SelfAvatar_ori/avatar/data/NeuMan/data/citron/masks/00000_0.png")
mask = mask > 0.5
mask = mask[:, :, 0]
image = cv2.imread("/home/ljr/Downloads/SelfAvatar_ori/avatar/data/NeuMan/data/citron/images/00000.png")
background_color = (0, 0, 0)
 
result = np.where(mask[..., np.newaxis], image, background_color)
result = result.astype(np.uint8)

# plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
# plt.axis('off')  # 关闭坐标轴
# plt.show()

seg_class = np.load("/home/ljr/Downloads/SelfAvatar_ori/avatar/data/NeuMan/data/citron/seg_class/00000.npy")
target_ids = [3,4,10,11,13,14,15,16,17]
binary_mask = np.isin(seg_class, target_ids).astype(np.uint8)
result0 = np.where(binary_mask[..., np.newaxis], image, background_color)
result0 = result0.astype(np.uint8)
plt.imshow(cv2.cvtColor(result0, cv2.COLOR_BGR2RGB))
plt.axis('off')  # 关闭坐标轴
plt.show()
```
